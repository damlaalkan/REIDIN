```python
import pandas as pd
import sys
import hashlib
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot

data = {'ID':["398732","398732","398732","398732","398732","398732","398732","398732","398732"],'Value':["7192617","7296132","7471138","7623971","8019314","8119122","5797402","6968042","7076301"],'Date':["2017-07-05","2017-07-31","2017-09-15","2017-10-28","2018-02-28","2018-04-12","2016-09-01","2017-05-12","2017-06-07"]}
df = pd.DataFrame(data)

print (df)

#linear regression does not work with date, convert to numerical value
import datetime as dt
df['Date'] = pd.to_datetime(df['Date'])
df['Date']=df['Date'].map(dt.datetime.toordinal)


y = np.asarray(df['Value'])
X = df[['Date']]

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.25,random_state=42)


model = LinearRegression() 
model.fit(X_train, y_train)
model.score(X_train, y_train) 
y_predict = model.predict(X_test)

print(y_predict)


df.plot()
pyplot.show()
```
